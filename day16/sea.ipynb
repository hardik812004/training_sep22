```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,10)
y=x**3
plt.plot(x,y)
# plt.show()
```

```python
plt.bar(x,y)
```

```python
plt.hist(x,y)
```

```python
plt.boxplot(x)
```

```python
plt.boxplot(y)
```

```python
plt.stem(x,y)
```

```python
plt.violinplot(y)
```

```python
plt.violinplot(x)
```

```python
plt.bar(x,y)
plt.title("Bar Graph")
plt.hist(x,y)
plt.title("Histogram")
```

```python
plt.bar(x,y)
plt.title("Bar Graph")
plt.show()
plt.hist(x,y)
plt.title("Histogram")
plt.show()
```

```python
plt.bar(x,y)
plt.title("Bar Graph")
plt.show()
plt.hist(x,y)
plt.title("Histogram")
plt.show()
plt.scatter(x,y)
plt.title("Scatter")
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')

plt.show()
```

```python
plt.scatter(x,y)
plt.title("Scatter")
plt.grid()
plt.xlabel('x-->')
plt.ylabel('y-->')
plt.xticks(x)
plt.yticks(range(0,200,10))
plt.show()
```

```python
y1=x**4
plt.plot(x,y,x,y1)
plt.legend(['y=x**3','y1=x**4'])
plt.show()
```

```python
fig=plt.figure()
plt.plot(x,y,lw=3,marker='o',markersize=15,markerfacecolor='yellow')
plt.plot(x,y1,marker='v',markersize=10)
plt.legend(['y=x**3','y1=x*4'])
plt.grid()
plt.show()
```
